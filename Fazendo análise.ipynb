```python
#Fazendo um mapa de densidade dos dados obtidos
```

```python
%%time
plt.figure(figsize=[10,6], dpi=700)
plt.suptitle("Estrelas no DR2", fontsize=12)
plt.hist2d(data.ra, data.dec, bins=50)
plt.xlabel("R.A. (deg)", fontsize=14)
plt.ylabel("Dec. (deg)", fontsize=14)
plt.colorbar(label="Densidade de Pontos")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
```

```python
plt.figure(figsize=[7,5])
plt.title("CMD - Estrelas no DES DR2", fontsize = 14)
plt.hexbin(data.gmr, data.mag_r, gridsize=500, bins='log') #essa linha cria um gráfico hexbin usando Matplot,
#que é conveniente para visualizar os dados de densidade de pontos em bidimensões
plt.xlabel("$g - r$", fontsize=14)
plt.ylabel("$g$", fontsize=14)
plt.colorbar(label="Densidade de Pontos")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(2.5,-1.5)
plt.ylim(24,16)
plt.tight_layout()
```
