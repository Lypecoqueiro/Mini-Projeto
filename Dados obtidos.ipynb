```python
dados = dados_1
del dados_1 #Limpando a memória dos dataframes
del dados_2 #não vamos utilizar
```

```python
new_columns = {"coadd_object_id": "object_id",
              "mag_auto_g_dered": "mag_g",
              "mag_auto_r_dered": "mag_r",
              "mag_auto_i_dered": "mag_i",
              "magerr_auto_g": "err_g",
              "magerr_auto_r": "err_r",
              "magerr_auto_i": "err_i"}
dados.rename(columns=new_columns, inplace=True)
```

```python
dados.info()
```

```python
dados.head()
```

```python
#cálculo da cor g-r (nova coluna no dataframe dados)
```

```python
dados["gmr"] = dados.mag_g - dados.mag_r
```

```python
dados.head()
```

```python
#Limpeza de dados e selecionando estrelas com fotometria de boa qualidade na banda g
#flags_g < 4 e com medidas bem sucedidas em magnitudes (mag diferentes de 99.)
```

```python
dados.query("flags_g < 4 & mag_g !=99. & mag_r !=99. ", inplace=True)
```

```python
dados.to_csv('dados_obtidos.csv', index=False)
```
